```python
import matplotlib.pyplot as plt

exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")

plt.axis([-2.5, 5, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')


ax.set_title("3D plot of Z = X^2 + Y^2")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

pokemon = pd.read_csv('./Pokemon.csv', encoding='latin1')

features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = pokemon[features]


pearson_corr = data.corr(method='pearson')


spearman_corr = data.corr(method='spearman')


plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', cbar=True)
plt.title('Pearson Correlation Heatmap')
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', cbar=True)
plt.title('Spearman Rank Correlation Heatmap')
plt.show()
```
